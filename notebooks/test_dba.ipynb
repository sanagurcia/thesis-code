```python
import numpy as np
import mydba
import ucrdata
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# get n sequences of same class
S = ucrdata.get_class_sequences(10)
```

```python
# calculate mean sample for n iterations
mean = mydba.dba_mean(S, 10)
```

```python
# plot sample sequences and mean sequence (black)
for i in range(S.shape[0]):
    plt.plot(S[i])

plt.plot(mean, color='black')
```
