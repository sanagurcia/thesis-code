```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# UCR Datasets: each line in a file contains a time series as list of CSVs
#  first entry of each sequences signifies target class (for clustering)

# Number of time series to visualize
n = 4

# Import entire dataset (450 sequences of 270 pts each)
words_file = open('UCR_TS_Archive_2015/50words/50words_TRAIN')

# initialize target array (rows=number of sequences, columns=points per sequence)
Y = np.zeros((n, 271)) 

for i in range(n):
    line = words_file.readline()              # read in one sequence
    Y[i] = np.array(line.split(','), float)   # transform comma separated string into np array
    
Y = Y[:,1:]     # remove first column (i.e. first entry from each time series)

# plot each time series
for i in range(Y.shape[0]):
    plt.plot(Y[i])
```
